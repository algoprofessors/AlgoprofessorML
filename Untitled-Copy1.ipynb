```python
reliance['DateTime']=reliance['Datetime'].apply(lambda x: datetime.datetime.strptime(str(x),'%Y-%m-%dT%H:%M:%S'))
```

```python
reliance['buy_signal_rsi']=0
reliance['sell_signal_rsi']=0
```

```python
intraday_trade_log = pd.DataFrame(columns=['Entry_Datetime','Entry_Price','Exit_Price','Exit_Datetime','PnL'])
position=False
```

```python
reliance['long']=np.where(reliance['RSI_21']<70,1,0)
reliance['short']=np.where(reliance['RSI_21']>30,-1,0)
reliance['position']=reliance['long']+reliance['short']
```

```python
position=True
```

```python
reliance[(reliance['long']==1)|(reliance['short']==1)]
```

```python

position=False
```

```python
k=0
for i in range(len(reliance)):
    if reliance['RSI_21'][i]>30 and position==False:
        reliance.loc[i,'buy_signal']=1
        entry_price=reliance['Close'][i]
        intraday_trade_log.loc[k,'Entry_Price']=reliance['Close'][i]
        intraday_trade_log.loc[k,'Entry_Datetime']=reliance['DateTime'][i]
        position=True
    if reliance['RSI_21'][i]<70 and position==True:
        reliance.loc[i,'sell_signal']=1
        exit_price=reliance['Close'][i]
        intraday_trade_log.loc[k,'Exit_Price']=reliance['Close'][i]
        pnl=exit_price-entry_price
        intraday_trade_log.loc[k,'Pnl']=pnl
        intraday_trade_log.loc[k,'Exit_Datetime']=reliance['DateTime'][i]
        k=k+1
        position=False
    
```

```python
reliance['return']=np.log(reliance['Close'] / reliance['Close'].shift(1) )
```

```python
new = pd.merge(reliance['Close'], reliance['RSI_21'], left_index=True, right_index=True)

#If the indicator’s line crosses the level 30 from below, a long position (Buy) is opened.  
new['long'] = np.where((new['RSI_21'] < 30),1,np.nan)
new['long'] = np.where((new['RSI_21'] > 70),0,new['long'])

new['long'].ffill(inplace=True)

new['gain_loss'] = new['long'].shift(1) * reliance['return']

new['total'] =  new['gain_loss'].cumsum()

print(new.tail(600))

#outcome
```
